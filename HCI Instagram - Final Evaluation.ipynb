```python
import speech_recognition as sr
import pyttsx3 as p
from selenium import webdriver
import time

r = sr.Recognizer()
engine = p.init()

def change_name():
    engine.say("I can hear that you want to change your name! Please type in the input box what you want me to call you to avoid the hassle!")
    engine.runAndWait()
    name = input("Enter your nickname here: ")
    f = open("instabotdata.txt","w")
    f.write(name)
    f.close()
    return name

def create_user():
    engine.say("Thank you for having me! I can see that you are a new user! Let us quickly setup your details!")
    engine.runAndWait()
    f = open("instabotdata.txt","w+")
    engine.say("What should I call you?")
    engine.runAndWait()
    mic = sr.Microphone()
    
    while 1:
        with mic as source:
            audio = r.listen(source, timeout=7, phrase_time_limit=7)
            try:
                text = r.recognize_google(audio)
                engine.say("Hello " + text + ". Whenever you need me just say hello Instagram.")
                f.write(text)
                f.close()
                return
            except sr.UnknownValueError:
                    print ("")

            except sr.RequestError as e:
                    print("")

def like(browser):
    lst = browser.find_elements_by_css_selector("[aria-label=Like]")
    i = 0
    while 1:
        try:
            if lst[i].is_displayed():
                lst[i].click()
                return
        except:
            i=i+1

def ask_commands(browser):
    engine.say("Let me know what you need!")
    engine.runAndWait()
    mic = sr.Microphone()
    height = 0
    while 1:
        with mic as source:
            print("Up for commands")
#             r.adjust_for_ambient_noise(source)
            audio=r.listen(source, timeout=5, phrase_time_limit=5)
            try:
                text=r.recognize_google(audio)
                print(text)
                if text=="scroll down":
                    print("Scrolling Down")
                    new_height = height + 1000
                    browser.execute_script("window.scrollTo(" + str(height) + "," + str(new_height) + ");")
                    height = new_height
                elif text == "scroll up":
                    print("Scrolling Up")
                    new_height = max(0,height - 1000)
                    browser.execute_script("window.scrollTo(" + str(height) + "," + str(new_height) + ");")
                    height = max(new_height,0)
                elif text == "scroll to top":
                    print("Scrolling Top")
                    new_height = 0
                    browser.execute_script("window.scrollTo(" + str(height) + "," + str(new_height) + ");")
                    height = 0
                elif text == "go home":
                    print("Going home")
                    height = 0
                    browser.get("https://www.instagram.com/")
                elif text == "like":
                    print("Liking this post!")
                    like(browser)
                elif text=="take care Instagram":
                    engine.say("I am happy to help! Have a good day.")
                    engine.runAndWait()
                    return
            except sr.UnknownValueError:
                    print ("")

            except sr.RequestError as e:
                    print("")
def instagram():
    browser = webdriver.Chrome('/Users/satty26/driver/chromedriver')
    link = 'https://www.instagram.com/accounts/login/'
    browser.get(link)
    time.sleep(5)
    
    fh = open('login.txt')
    fh = fh.readlines()
    username = fh[0].strip()
    password = fh[1].strip()
    
    elementID = browser.find_elements_by_xpath("//*[@id='loginForm']/div/div[1]/div/label/input")[0]
    elementID.send_keys(username)
    elementID = browser.find_elements_by_xpath("//*[@id='loginForm']/div/div[2]/div/label/input")[0]
    elementID.send_keys(password)
    elementID.submit()
    time.sleep(5)
    
    try:
        browser.find_elements_by_xpath("//*[@id='react-root']/section/main/div/div/div/div/button")[0].click()
    except:
        pass
    time.sleep(5)
    try:
        browser.find_elements_by_xpath("/html/body/div[5]/div/div/div/div[3]/button[2]")[0].click()
    except:
        pass
    engine.say("We are good to go!")
    engine.runAndWait()
    ask_commands(browser)
    browser.quit()

try:
    f = open("instabotdata.txt")
except:
    create_user()

file = open("instabotdata.txt", "r")
nick_name =  file.read()

engine.say("I am glad to meet you " + nick_name + " . Whenever you need me just say hello Instagram. If you would like to change your name, please say change my name.")
engine.runAndWait()

# user_details = setup_user
mic = sr.Microphone()
while 1:
    with mic as source:
        print("Up")
#         r.adjust_for_ambient_noise(source)
        audio=r.listen(source, timeout=10,  phrase_time_limit=10)
        try:
            text=r.recognize_google(audio)
            print(text)
            if text=="hello Instagram":
                engine.say("Hello " + nick_name + ", I am logging in to your instagram account!")
                engine.runAndWait()
                instagram()
                break
            elif text=="change my name":
                nick_name = change_name()
            
        except sr.UnknownValueError:
                print ("")

        except sr.RequestError as e:
                print("")

# print()
# engine.say("What can I do for you?")    
# engine.runAndWait()
        
```
